```python
import pandas as pd;
import numpy as np;
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error
```

```python
import matplotlib.pyplot as plt
```

```python
diabetes=datasets.load_diabetes()
```

```python
print(diabetes.data)
```

```python
diabetes_x=diabetes.data[:,np.newaxis,2]
diabetes_x
```

```python
diabetes_xtrain=diabetes_x[:-30]
diabetes_xtest=diabetes_x[-30:]
diabetes_ytrain=diabetes.target[:-30]
diabetes_ytest=diabetes.target[-30:]
```

```python
model=linear_model.LinearRegression()
print(diabetes_xtest)
```

```python
model.fit(diabetes_xtrain,diabetes_ytrain);
```

```python
prediction=model.predict(diabetes_xtest);
```

```python
print(mean_squared_error(diabetes_ytest,prediction))
```

```python
print(model.coef_,model.intercept_)
```

```python
plt.scatter(diabetes_xtest,diabetes_ytest)
```

```python
plt.scatter(diabetes_xtest,diabetes_ytest)
plt.plot(diabetes_xtest,prediction)
```
