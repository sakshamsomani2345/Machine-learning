```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y=x**2
y2=2*x+3
print(x)
print(y)
print(y2)
```

```python
plt.plot(x,y)
plt.show()
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x,y,color="green")
plt.show()
```

```python
plt.plot(x,y,color="red",label="guava",marker="o")
plt.plot(x,y2,label="apple",linestyle="dashed")
plt.title("this is a graph")
plt.xlabel("time")
plt.legend()
plt.show()
```
